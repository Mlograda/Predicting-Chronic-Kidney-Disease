## Diagnose missing values

In this section, I'm going to diagnose the type of the missing values in the dataset. (i.e., is the missing values o MCAR, MAR, or MNAR type).



```python
# HELP FUNCTIONS TO DIAGNOSE THE MISSING VALUES
# Writing a function to diagnose the missing values both visually and statistically (numerical attrbiutes)
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}
    
    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())    
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])
   
    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()
    
    plt.figure(figsize=(10,4))
    
    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()
    
    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()
    
    p_value = ttest_ind(group_1_data,group_2_data).pvalue
    
    print('p-value of t-test: {}'.format(p_value))
    
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(14,6))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])   
    p_value = chi2_contingency(contigency_table)[1]
    
    print('p-value of Chi_squared test: {}'.format(p_value))
```

I'm going to evaluate whether the value of a numerical attribute is significantly different among the data objects having missing values under rc attribute and the data objects without missing values under the rc attribute. 

```python
# Diagnosing the MVs of all the attributes present in the dataset visually and statistically.

# Get the numerical attrbiutes

numerical_attributes = ['sc', 'pcv', 'sg', 'hemo']

categorical_attributes = ['diagnosis']

# create the boolean mask of missing values
BM_MV = kidney_df.rc.isna()

for att in numerical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Numerical(kidney_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')

for att in categorical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Categorical(kidney_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

diagnosing missing vaues in rc based on the other attributes:
* sc : not significantly diffrent 
* pcv : significantly different
* sg : significantly different
* hemo: significantly different 
* diagnosis: significantly different

conclusion : missing values in rc are of (MAR) type

The tendency of all of the MAR missing values is highly linked to the diagnosis dependent attribute.


I'm going to use the Decision Tree algorithm to predict the dependent variable ('diagnosis'). 
Through the diagnosing of missing values. I realized that the tendancy of some of the missing values can be a predictor of the dependent attribute.  This is important because if we were to impute the missing values, that would remove this valuable information from the dataset; the valuable information is that the missing values of some of the attributes (the MAR ones) predict the dependent attribute. Therefore, regardless of the imputation method that we will use, we will add a binary attribute to the dataset for every attribute with MAR missing values that describes whether the attribute had a missing value. These new binary attributes will be added to the independent attributes of the classification task to predict the diagnosis dependent attribute.


```python
# introducing binary attributes to the dataset
kidney_df['rc_BMV'] = kidney_df.rc.isna().astype(int)
kidney_df['pcv_BMV'] = kidney_df.pcv.isna().astype(int)
kidney_df['sg_BMV'] = kidney_df.sg.isna().astype(int)
kidney_df['hemo_BMV'] = kidney_df.hemo.isna().astype(int)
```

```python
kidney_df
```

Decision Tree algorithm consecutively splits data objects into groups based on the value of the attributes, and when the data objects have values that are larger than or smaller than the central tendencies of the attribute, they are more likely to be classified with a specific label. Therefore, by imputing with the central tendency of the attributes, we will not introduce a bias into the dataset, so the imputed value will not cause the classifier to predict one label over the other more often.

**The question that we now need to answer is: Which central tendency should we use—median or mean?**

I need to check whether there are loads of outliers in the independent variables 



```python
kidney_df.describe()
```

```python

numerical_attributes = ['pcv','rc','sc','sg','hemo']
for att in numerical_attributes:
    kidney_df[att].plot.box(vert=False)
    plt.show()
```

After investigating the boxplots of the attributes with missing values, you will see that sc has too many outliers, and the rest of the attributes are not highly skewed. Therefore, sc will be imputed by the median() of the sc population and the MVs in th other attributes will be imputed with the mean() of each attribute population. 


```python
# imputing sc

kidney_df.sc.fillna(kidney_df.sc.median(), inplace=True)

kidney_df.fillna(kidney_df.mean(), inplace=True)
```

```python
kidney_df.info()
```

```python
ckd_df = kidney_df
```

```python
# building the decision tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
predictors = ['rc', 'sc', 'pcv', 'sg', 'hemo', 'rc_BMV', 'pcv_BMV', 'sg_BMV', 'hemo_BMV']
target = 'diagnosis'

Xs = ckd_df[predictors]
y= ckd_df[target]

classTree = DecisionTreeClassifier(min_impurity_decrease= 0.01, min_samples_split= 15)
classTree.fit(Xs, y)
```

```python

from sklearn.tree import plot_tree
plt.figure(figsize=(15,15))
plot_tree(classTree, 
          feature_names=predictors,  
          class_names=y.unique(),
          filled=True,
          impurity=False)
plt.show()
```
